```python
from pandas.plotting import andrews_curves

#here we load again the data because we need a column with the spacies name

data = pd.read_csv(r'/data/shared/Big-Geodata-Processing-2021-1A/distributed_computing_machine_learning/Data/Iris/iris.dataComma.txt',
                   header = None,
                   names = ['sepal length','sepal width','petal length','petal width','species'])

plt.figure(figsize = (7, 7),
          dpi = 133)

andrews_curves(data,'species', samples = 500)

plt.show()
```

```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize = (7, 7))
parallel_coordinates(data, 'species')

plt.savefig(r'private/Tutorial/parallel.png', format = 'png')

plt.show()
```
